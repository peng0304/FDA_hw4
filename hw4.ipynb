```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn import linear_model,metrics
from sklearn.preprocessing  import StandardScaler
```

# Data discription
- Each row of the dataset is a single game with the following features (in the order in the vector):
- Team won the game (1 or -1)
- Cluster ID (related to location)
- Game mode (eg All Pick)
- Game type (eg. Ranked)
- end: Each element is an indicator for a hero. Value of 1 indicates that a player from team '1' played as that hero and '-1' for the other team. Hero can be selected by only one player each game. This means that each row has five '1' and five '-1' values.

# Problem
- 判斷這個隊伍有沒有獲勝(第一個Attribute為1)

# data preprocessing
- 利用Cluster ID、遊戲模式、遊戲類型去判斷隊伍輸或贏

```python
df_train = pd.read_csv('./dota2Train.csv')
df_test = pd.read_csv('./dota2Test.csv')
```

```python
train_X = df_train.iloc[:, 1:4].reset_index(drop=True)
train_X.head()
```

```python
train_Y = df_train.iloc[:,[0]]
train_Y.head()
```

```python
test_X = df_test.iloc[:, 1:4].reset_index(drop=True)
test_X.head()
```

```python
test_Y = df_test.iloc[:,[0]]
test_Y.head()
```

```python
sc = StandardScaler()
sc.fit(train_X)

x_train=sc.transform(train_X)
x_test=sc.transform(test_X)
```

```python
mlp = MLPClassifier(solver='lbfgs', alpha=5e-2,hidden_layer_sizes=(5, 2), random_state=1)
mlp.fit(x_train, train_Y)
mlp.score(x_test, test_Y)
```

# Result
- Training的結果為53.5%

# Data preprocessing2
- 利用所選取的英雄去判斷輸贏

```python
train_X = df_train.iloc[:, 4:].reset_index(drop=True)
train_Y = df_train.iloc[:,[0]]
train_X.head()
```

```python
test_X = df_test.iloc[:, 4:].reset_index(drop=True)
test_Y = df_test.iloc[:,[0]]
test_X.head()
```

```python
sc = StandardScaler()
sc.fit(train_X)

x_train=sc.transform(train_X)
x_test=sc.transform(test_X)
```

```python
mlp = MLPClassifier(solver='lbfgs', alpha=5e-2,hidden_layer_sizes=(5, 2), random_state=1)
mlp.fit(x_train, train_Y)
mlp.score(x_test, test_Y)
```

# Result
- Training結果為59.3%

# Conclusion 
- 我分成兩類，一類為Cluster ID、對戰模式以及遊戲形態，另一類為英雄對戰組合。
- 我試過將Cluster ID獨立出來後Training，結果不變，因此可看出在第一種data preprocessing下，主要控制的因素在於對戰模式以及遊戲形態，與玩家無關。
- 第二種的準確度比較高，表示對戰組合導致的勝負結果會比較準，但準確性也才59.3%。

# More information
- 從這個資料裡還可以得到更多的資訊，例如在這個賽季中最常被使用的英雄、或是最常出現的對戰組合，進而分析強勢的英雄種類。
- 相反地，根據出場率最低的英雄，可讓遊戲公司去做角色的微調。

```python
n = 0
freqency = []
for i in range(4,117):
    for j in range(0,10000):
        if df_train.iloc[j,i]==1:
            n = n+1
    freqency.append(n)
    n=0
print(freqency)
```

```python
print(np.where(freqency==np.max(freqency)))
print(np.where(freqency==np.min(freqency)))
```

# Result
- 第9號英雄出場率為最高，因此可能為此賽季的強勢角色。
- 第24、108號英雄出場率最低，因此為弱勢角色，遊戲公司可參考並調整此角色。

# Improvement
- 藉由強勢英雄，可以去分析當隊伍拿到強勢英雄時，勝率應該會提高，加入這項attribute再分析，讓準確度提高。

```python
top_hero=[]
for i in range(0,92649):
    if df_train.iloc[i,12]==1:
        top_hero.append(1)
    else:
        top_hero.append(0)
df_train['top']= top_hero
df_train.head()
```

```python
top_hero=[]
for i in range(0,10293):
    if df_test.iloc[i,12]==1:
        top_hero.append(1)
    else:
        top_hero.append(0)
df_test['top']= top_hero
df_test.head()
```

```python
train_X = df_train.iloc[:, 1:].reset_index(drop=True)
train_Y = df_train.iloc[:,[0]]
train_X.head()
```

```python
test_X = df_test.iloc[:, 1:].reset_index(drop=True)
test_Y = df_test.iloc[:,[0]]
test_X.head()
```

```python
sc = StandardScaler()
sc.fit(train_X)

x_train=sc.transform(train_X)
x_test=sc.transform(test_X)
```

```python
mlp = MLPClassifier(solver='lbfgs', alpha=5e-2,hidden_layer_sizes=(5, 2), random_state=1)
mlp.fit(x_train, train_Y)
mlp.score(x_test, test_Y)
```
